# สร้าง Model

```python
# แบ่งข้อมูลเป็น X และ y
X, y = cs2.drop('upgrade', axis=1), cs2[['upgrade']]
```

```python
# แสดงตัวอย่างข้อมูล X
X.head()
```

```python
# แสดงตัวอย่างข้อมูล y
y.head()
```

แปลงข้อมูล categorical variables ให้เป็นตัวเลขโดยให้เลือกใช้ OneHotEncoder จาก category_encoder package.

```python
catfeat = ['สิทธิ','จำนวนวัน','จังหวัด','ตำบล','เพศ']
```

```python
!pip install category_encoders
```

```python
import category_encoders as ce

# แปลงข้อมูล categorical เป็นตัวเลข
ce_ONE = ce.OneHotEncoder(cols=catfeat)
encoded_X = ce_ONE.fit_transform(X)
```

```python
# แสดงตัวอย่างข้อมูล X ที่แปลงแล้ว
encoded_X.head()
```

แบ่งข้อมูลที่แปลงได้ออกเป็น training และ testing set

```python
from sklearn.model_selection import train_test_split

# แบ่งข้อมูลเป็น training และ testing set
X_train, X_test, y_train, y_test = train_test_split(encoded_X, y, random_state=0)
```

สร้างโมเดลด้วย Logistic Regression, Linear support vector machines และ k-nearest neighbors และประเมินผลด้วย training set

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier 
```

```python
lr = LogisticRegression().fit(X_train, y_train)
svm = LinearSVC().fit(X_train, y_train)
knn = KNeighborsClassifier().fit(X_train, y_train)
```

```python
print('Logistic Regression training score: {:.3f}'.format(lr.score(X_train, y_train)))
print('Linear SVC training score: {:.3f}'.format(svm.score(X_train, y_train)))
print('kNN training score: {:.3f}'.format(knn.score(X_train, y_train)))

# test score ของ Logistic Regression มีค่ามากที่สุด 
# รองลงมาคือ kNN และ Linear SVC ตามลำดับ
```

ปรับค่า parameters ด้วยการใช้ GridSearchCV แล้วประเมินผลด้วยข้อมูล test set

```python
from sklearn.model_selection import GridSearchCV
```

```python
# หา best estimator ของ Logistic Regression
param = {'penalty':['l1','l2'], 'C':[0.001,0.01,0.1,1,10,100,1000]}
lr = LogisticRegression()
lr_grid = GridSearchCV(lr, param)
lr_grid.fit(X_train, y_train)

print('GridSearchCV-Logistic Regression')
print('Best Penalty: ', lr_grid.best_estimator_.get_params()['penalty'])
print('Best C: ', lr_grid.best_estimator_.get_params()['C'])
```

```python
# หา best estimator ของ Linear SVC
param = {'penalty':['l1','l2'], 'C':[0.001,0.01,0.1,1,10,100,1000]}
svc = LinearSVC()
svc_grid = GridSearchCV(svc, param)
svc_grid.fit(X_train, y_train)

print('GridSearchCV-LinearSVC')
print('Best Penalty: ', svc_grid.best_estimator_.get_params()['penalty'])
print('Best C: ', svc_grid.best_estimator_.get_params()['C'])
```

```python
# หา best estimator ของ kNN
param = {'n_neighbors':[1,3,5,7,9,11]}
knn = KNeighborsClassifier()
knn_grid = GridSearchCV(knn, param)
knn_grid.fit(X_train, y_train)

print('GridSearchCV-kNN')
print('Best Penalty: ', knn_grid.best_estimator_.get_params()['n_neighbors'])
```

```python
# แสดง test score หลังปรับค่า parameters
lr = LogisticRegression(penalty='l2', C=0.001).fit(X_train, y_train)
svm = LinearSVC(penalty='l2', C=0.001).fit(X_train, y_train)
knn = KNeighborsClassifier(n_neighbors=1).fit(X_train, y_train)

print('Logistic Regression test score: {:.3f}'.format(lr.score(X_test, y_test)))
print('Linear SVC test score: {:.3f}'.format(svm.score(X_test, y_test)))
print('kNN test score: {:.3f}'.format(knn.score(X_test, y_test)))

# test score ของ Logistic Regression มีค่ามากที่สุด
# รองลงมาคือ Linear SVC และ kNN ตามลำดับ
```
